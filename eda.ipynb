```python
#importing necessary libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import plotly.express as px
import seaborn as sns



df = pd.read_csv('G:\\Python\\.venv\\ICRISAT-District Level Data - ICRISAT-District Level Data.csv')
df
```

```python
#Data Cleaning

df.isnull().sum() #checks for missing values in each column
```

```python
#checks for special(-1) values in the DataFrame

(df==-1).sum() 
```

```python
#Replacing -1 values with NaN

df=df.replace(-1, np.nan) #replaces -1 values with NaN
```

```python
#Fill Nan values with the mean

df=df.fillna(df.mean(numeric_only=True)) 
```

```python
#Checks for null values again after cleaning

df.isna().sum()
```

```python
# Cleans the column names by stripping whitespace, converting to lowercase, removing units in parentheses, and replacing spaces with underscores

df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(r"\(.*?\)", "", regex=True)  # remove units
    .str.replace(" ", "_")
    .str.rstrip("_")  # remove trailing underscores
)
```

```python
# Display the cleaned column names

df.columns
```

```python
#Removing columns and rows with all NaN values
df = df.dropna(axis=1, how='all')
df = df.dropna(axis=0, how='all')
```

```python
#Checks for null values in the DataFrame
df.isnull().sum()
```

```python
df.duplicated().sum() #checks for duplicate rows in the DataFrame
```

```python
#Impact of Area Cultivated on Production (Rice, Wheat, Maize)
#Scatter plots with regression lines to show the relationship between area cultivated and production for Rice, Wheat


fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.regplot(x='rice_area', y='rice_production', data=df, ax=axes[0], scatter_kws={'alpha':0.3},color='darkgreen')
axes[0].set_title("Rice: Area vs Production")

sns.regplot(x='wheat_area', y='wheat_production', data=df, ax=axes[1], scatter_kws={'alpha':0.3}, color='purple')
axes[1].set_title("Wheat: Area vs Production")

sns.regplot(x='maize_area', y='maize_production', data=df, ax=axes[2], scatter_kws={'alpha':0.3}, color='red')
axes[2].set_title("Maize: Area vs Production")

plt.tight_layout()
plt.show()
```
